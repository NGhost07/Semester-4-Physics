# New section

Import library for model

```python
import matplotlib.pyplot as plt
import numpy as np
```

Set up initial value

```python
# Constants
pi = np.pi
# Постоянная Планка
h = 6.63e-34
# Масса Электрона 
m = 9.1e-31 
# ширина ямы
L = 2 

#Values for A = sqrt(2/L) - это амплитуда колебаний and x
xvals = np.linspace(0,L,100)
A =(2/L) ** (1/2) 
```

Волновые функции частицы в яме

```python
def f(n,x):
  wf = A*np.sin(n*pi*x/L)
  return wf
```

Энергия функции

```python
def E(n):
  return ((pi*h*n)**2 / (2*m*L**2))
```

График для первых 5 состояний

```python
yvals1 = f( 1, xvals)
yvals2 = f( 2, xvals)
yvals3 = f( 3, xvals)
yvals4 = f( 4, xvals)
yvals5 = f( 5, xvals)

plt.figure(figsize=(15,12))
plt.plot(xvals, yvals1, label = "E_1 = {}".format(E(1)))
plt.plot(xvals, yvals2, label = "E_2 = {}".format(E(2)))
plt.plot(xvals, yvals3, label = "E_3 = {}".format(E(3)))
plt.plot(xvals, yvals4, label = "E_4 = {}".format(E(4)))
plt.plot(xvals, yvals5, label = "E_5 = {}".format(E(5)))

plt.legend()
plt.title("График для первых 5 состояний")
plt.xlabel("x")
plt.ylabel("ψ")
plt.grid()
plt.show()
```
